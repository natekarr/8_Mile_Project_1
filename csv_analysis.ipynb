```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# File to Load
csv_path = Path("Resources/mxmh_survey_results.csv")

# Read the modified Comic Books csv and store into Pandas DataFrame
data = pd.read_csv(csv_path) #encoding="utf-8"
df=pd.DataFrame(data)
df
```

```python
df.columns
```

```python
df['Primary streaming service'].describe()
```

```python
df.dtypes
```

```python
df=df.astype({"Frequency [Classical]": str}, errors='raise')
df=df.astype({"Frequency [Country]": str}, errors='raise') 
df=df.astype({"Frequency [EDM]": str}, errors='raise')
df=df.astype({"Frequency [Folk]": str}, errors='raise')
df=df.astype({'Frequency [Gospel]': str}, errors='raise')
df=df.astype({"Frequency [Hip hop]": str}, errors='raise')
df=df.astype({"Frequency [K pop]": str}, errors='raise')
df=df.astype({"Frequency [Latin]": str}, errors='raise')
df=df.astype({"Frequency [Lofi]": str}, errors='raise')
df=df.astype({"Frequency [Metal]": str}, errors='raise')
df=df.astype({"Frequency [Pop]": str}, errors='raise')
df=df.astype({"Frequency [Jazz]": str}, errors='raise')
df=df.astype({"Frequency [R&B]": str}, errors='raise')
df=df.astype({"Frequency [Rap]": str}, errors='raise')
df=df.astype({"Frequency [Rock]": str}, errors='raise')
df=df.astype({"Frequency [Video game music]": str}, errors='raise')
```

```python
mapping={'Never': 0,
        'Rarely': 1,
        'Sometimes':2, 
        'Very frequently': 3}

df= df.replace({'Frequency [Classical]': mapping,
'Frequency [Country]': mapping, 
'Frequency [EDM]': mapping, 
'Frequency [Folk]': mapping,
'Frequency [Gospel]': mapping, 
'Frequency [Hip hop]': mapping, 
'Frequency [Jazz]': mapping,
'Frequency [K pop]': mapping,
'Frequency [Latin]': mapping,
'Frequency [Lofi]': mapping,
'Frequency [Metal]': mapping,
'Frequency [Pop]': mapping,
'Frequency [R&B]': mapping,
'Frequency [Rap]': mapping,
'Frequency [Rock]': mapping,
'Frequency [Video game music]': mapping})
    
df.head()
```

```python
df=df[(df['Primary streaming service'] =="Spotify")]
df
```

```python
x_axis=df["Music effects"].astype(str)
y_axis=df["Frequency [R&B]"]

plt.bar(x_axis, y_axis)
plt.title("Frequency of R&B and Music Effects on Mental Health")
plt.xlabel("Effect")
plt.ylabel("Frequency of Listening")
plt.show()
```
